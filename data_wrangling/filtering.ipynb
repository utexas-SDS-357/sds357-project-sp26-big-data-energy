```python
# Import libraries
import pandas as pd
import geopandas as gpd
```

```python
# Load data
df = pd.read_csv('/Users/inhae/Downloads/san_antonio_encoded.csv')
original_count = len(df)
```

```python
# Convert stop data to GeoDataFrame
stops = gpd.GeoDataFrame(
    df, geometry=gpd.points_from_xy(df["lng"], df["lat"]),crs="EPSG:4326")
```

```python
# Load San Antonio boundary
boundary = gpd.read_file("/Users/inhae/Downloads/COSABoundary.geojson")
```

```python
# Fix CRS mismatch in boundary file
boundary = boundary.set_crs(epsg=3857, allow_override=True)
boundary = boundary.to_crs(epsg=4326)
```

```python
# Dissolve into one polygon
boundary_union = boundary.dissolve()
polygon = boundary_union.geometry.iloc[0]
```

```python
# Keep only points inside or on the boundary
stops_filtered = stops[stops.geometry.intersects(polygon)].copy()
```

```python
# Drop geometry and save filtered dataset
stops_filtered.drop(columns="geometry").to_csv(
   "sa_policy_filtered.csv",
   index=False
)

print("Original rows:", original_count)
print("Rows after spatial filtering:", len(stops_filtered))
```
