```python
import pandas as pd

san_antonio = pd.read_csv('/Users/arjunrajesh/Desktop/new2_san_antonio.csv')
categorical_cols = ['subject_race', 'subject_sex', 'outcome', 'type']

#san_antonio.head()
# One-hot encoding with integer output (1 and 0)
san_antonio_data_encoded = pd.get_dummies(san_antonio, columns=categorical_cols, drop_first=True, dtype=int)

# Verify the first few rows
#print(data_encoded.filter(like='raw_race').head())
print(san_antonio_data_encoded.filter(like='type').head())

```

```python
# List your boolean columns
bool_cols = ['arrest_made', 'citation_issued', 'search_conducted']

# Convert them to 1 (True) and 0 (False)
san_antonio_data_encoded[bool_cols] = san_antonio_data_encoded[bool_cols].astype(int)

# Verify the change
print(san_antonio_data_encoded[bool_cols].head())
```

```python
#Drop the subject_race column
san_antonio_data_encoded.drop('raw_race', axis=1, inplace=True, errors='ignore')

#Verify it is gone
print("raw_custodial_arrest_made" in san_antonio_data_encoded.columns)
```

```python

# Create new column
san_antonio_data_encoded["extreme_cases"] = ((san_antonio_data_encoded["arrest_made"] == 1) | 
                       (san_antonio_data_encoded["search_conducted"] == 1)).astype(int)
```
